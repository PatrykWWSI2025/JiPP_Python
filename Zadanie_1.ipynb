### Zmienne tekstowe (string)

```python
# Utwórz zmienną tekstową "name" i przypisz jej na wartość "Bob"
name = 'Bob'
```

```python
# Wydrukuj wartość zmiennej "name"
print(name)
```

```python
# Sprawdź typ danych zmiennej "name"
type(name)
```

```python
# Konkatenacja (łączenie) dwóch ciągów znaków

a = "Learning"
b = "Python"

# Połącz ciąg A i B

c = a + b # Operatora „+” można użyć do połączenia dwóch ciągów
print(c)
```

```python
# Konkatenacja (łączenie) dwóch ciągów znaków

a = "Learning"
b = "Python"

# Połącz ciąg A i B: „+” Operator może być używany do połączenia dwóch ciągów

c = a +" "+ b 
print(c)
```

```python
# Cięcie łańcucha znaków (stringa)

a = "Welcome to Python Programming"

# Wyciągnij słowo Python - indeksowanie znaków zaczyna się od pozycji  0

print(a[11:17])
```

```python
# Formatowanie ciągu znaków

slices = 2
item = 'cake'
price = 50

text = "I need {} slices of {} for {} dollars!"
order = text.format(slices,item,price)
print(order)
```

```python
# Zapytaj użytkownika o nazwisko

name = input("What is your name?: ")

# Zapytaj użytkownika o wiek
age = input("How old are you?")

# Zapytaj użytkownika o miasto
city = input("Type city")

# Zapytaj użytkownika o jego zainteresowania
interests = input("What are your interests?")

# Utwórz tekst wyjściowy za pomocą formatowania ciągów

string = "Hi {name}, you are {age} years old. You live in {city} and you love {interests}!"

output = string.format(name=name, age=age, city=city, interests=interests)

# Wydrukuj tekst wyjściowy
print(output)
```

### Zmienne logiczne (boolean)

```python
# Wyrażenie porównawcze: (10 > 9)
10 > 9
```

```python
# Przypisz zmiennej "result" rezultat powyższego wyrażenia porównawczego
result = 10 > 9
```

```python
# Wydrukuj wartość zmiennej "result"
print(result)
```

```python
# Sprawdź typ danych zmiennej "result"
type(result)
```
