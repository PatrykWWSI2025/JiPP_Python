```python
# Utw√≥rz zmienne "vowels" (samog≈Çoski) i "consonants" (sp√≥≈Çg≈Çoski) i przypisz ka≈ºdej z nich warto≈õƒá 0
vowels = 0
consonants = 0
vowel_letters = "aeiouyƒÖƒô√≥AEIOUYƒÑƒò√ì"

# Utw√≥rz pƒôtlƒô i przeiteruj ≈Ça≈Ñcuch znak√≥w ‚ÄûProgramowanie Pythona‚Äù
for char in "Programowanie Pythona":
    


# Utw√≥rz instrukcjƒô warunkowƒÖ IF-ELSE, kt√≥ra wyliczy liczbƒô samog≈Çosek i sp√≥≈Çg≈Çosek w danym ≈Ça≈Ñcuchu znak√≥w
    if char.isalpha():  # Sprawdza, czy znak jest literƒÖ
        if char in vowel_letters:
            vowels += 1
        else:
            consonants += 1
    
        
# Wydrukuj ≈ÇƒÖcznƒÖ liczbƒô samog≈Çosek i sp√≥≈Çg≈Çosek w danym ≈Ça≈Ñcuchu znak√≥w
print("Liczba samog≈Çosek:", vowels)
print("Liczba sp√≥≈Çg≈Çosek:", consonants)
```

```python
# Tworzymy symulator ciekawskiego dziecka za pomocƒÖ pƒôtli


#Program SymulujƒÖcy Ciekawskiego Malucha

#Wszyscy wiemy, ≈ºe dzieci sƒÖ z natury ciekawe i ciƒÖgle zadajƒÖ pytania. Czƒôsto nie przekonujƒÖ ich odpowiedzi, wiƒôc wciƒÖ≈º pytajƒÖ: dlaczego?

#Spr√≥bujmy to odtworzyƒá i zbudowaƒá program w Pythonie, kt√≥ry symuluje zachowanie ciekawego malucha przy u≈ºyciu pƒôtli while‚Ä¶

#Maluch bƒôdzie ciƒÖgle powtarza≈Ç ‚Äûdlaczego‚Äù, dop√≥ki rodzic nie powie ‚Äûto wszystko!‚Äù

#Zatem spr√≥bujmy zbudowaƒá logikƒô opartƒÖ na tym, u≈ºywajƒÖc pƒôtli while.
answer = ""

while answer.lower() != "to wszystko!":
    print("Dlaczego?")
    answer = input("Rodzic odpowiada: ")

print("Dobrze, ju≈º nie pytam üôÇ")
```

```python
# U≈ºyj modu≈Çu random
import random

# Stw√≥rz listƒô "questions" sk≈ÇadajƒÖcƒÖ siƒô z 3 pyta≈Ñ, kt√≥re czƒôsto zadajƒÖ dzieci

questions = ["Why is the sky blue?","Why is the sun round?","Where are all the dinosaurs?"]

# Wybierz losowe pytanie z danej listy za pomocƒÖ instrukcji warunkowej
question = random.choice(questions)


# Zadaj wybrane pytanie za pomocƒÖ funkcji input()
# Pytania muszƒÖ zachowaƒá jednolite formatowanie
# Aby to uzyskaƒá, przekonwertuje wszystkie odpowiedzi na ma≈Çe litery i usu≈Ñ wszelkie nadmiarowe spacje
answer = ""

while answer != "to wszystko":
    print(question)
    answer = input(question).strip().lower()
    print("Rodzic odpowiada:",answer)
# Poczekaj do czasu, a≈º u≈ºytkownik wprowadzi has≈Ço ‚ÄûTo wszystko‚Äù


    
# Wy≈õwietl wiadomo≈õƒá
print("Dobrze, koniec pyta≈Ñ üôÇ")
```

```python
# Zmienny zakres funkcji

x = 98 # Globalny zakres zmiennej

def f1():
    global x
    x = 50 
    print(x)
    
def f2():
    x = 30 # zmienna lokalna
    print(x)
    
f1()
f2()
print(x)
```

```python
# Tworzenie funkcji bez domy≈õlnych parametr√≥w

def intro(name,city,interest):
    intro = 'Hi I am {},I live in {} and I love {}'.format(name,city,interest)
    return intro

intro("Alice","Mumbai","Music")

intro("Ryan","Delhi") # Brak argumentu zwr√≥ci b≈ÇƒÖd
```
